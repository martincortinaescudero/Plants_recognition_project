```python
import pandas as pd
import matplotlib.pyplot as plt

# Data from the table
data = {
    "The US": [78, 10, 12],
    "Germany": [12, 72, 15],
    "Mexico": [0, 10, 90]
}

# Index for the categories
index = ["Farming", "Industry", "Raw materials"]

# Creating DataFrame
df = pd.DataFrame(data, index=index)

# Transposing the DataFrame to have countries on the x-axis and activities as colors
df_transposed = df.T

# Plotting the bar graph with countries on the x-axis and custom colors
ax = df_transposed.plot(kind="bar", figsize=(10, 6), width=0.8, color=["#bcbd22", "#1f77b4", "#ff9900"])

# Adding labels and title
plt.ylabel("Percentage", fontweight='bold')
plt.xlabel("Country", fontweight='bold')
plt.xticks(rotation=0)

# Adding values on top of the bars
for container in ax.containers:
    for bar in container:
        height = bar.get_height()
        ax.annotate(f'{height}%',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Adding a legend outside the plot area
ax.legend(title='Categories', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjusting layout to make room for the legend
plt.tight_layout()

# Display the plot
plt.show()
```
