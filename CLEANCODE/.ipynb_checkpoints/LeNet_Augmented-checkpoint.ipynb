```python
X = []
Y = []
width = 200
height = 200

# resize images
for image, label in images_data:
    image = image.resize((width, height))
    X.append(np.array(image))
    Y.append(label)
```

```python
# Get all unique categories
unique_categories = set(Y)

# Create a mapping from category name to integer label
category_to_label = {category: label for label, category in enumerate(unique_categories)}

# Create a list to store the integer labels corresponding to each category
Y_labels = [category_to_label[category] for category in Y]

# Count the number of images in each category
category_counts = collections.Counter(Y)

# Print the mapping of category names to integer labels
print("Category to Label Mapping:")
for category, label in category_to_label.items():
    count = category_counts[category]
    print(f"{category} --> {label} --> {count} images")

X = np.array(X)
Y_labels = np.array(Y_labels)
```

# Definir los factores para el filtro verde dbil y el umbral para el filtro blanco intenso
faktor_green = 1.4
threshold_white = 150

# Crear una lista para almacenar las imgenes filtradas
X_filtrada = []

# Filtrar cada imagen en X y almacenar las imgenes filtradas en X_filtrada
for image in X:
    # Aplicar la mscara para filtrar colores verdes y blancos
    mask_green = (image[:, :, 1] < faktor_green * image[:, :, 0]) & (image[:, :, 1] < faktor_green * image[:, :, 2])
    mask_white = (image[:, :, 0] > threshold_white) & (image[:, :, 1] > threshold_white) & (image[:, :, 2] > threshold_white)
    # Combinar las mscaras utilizando OR
    mask = mask_green | mask_white  
    # Convertir a negro aquellos pxeles donde verde es dbil o blanco es intenso
    image[mask] = [0, 0, 0]  

    # Agregar la imagen filtrada a la lista X_filtrada
    X_filtrada.append(image)

# Sobrescribir X con las imgenes filtradas
X = np.array(X_filtrada)


```python
# Train and testing split (70/30)
X_train, X_test, y_train, y_test = train_test_split(X,Y_labels,test_size=0.3,random_state=42,shuffle=True, stratify=Y)
```

```python
X_train = X_train / 255
X_test = X_test / 255

y_test_no_categorical = y_test

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

# Second part: Data augmentation

```python
# Implement under the name train_datagen an image generator.
# Implement under the name test_datagen a test generator.

train_datagen = ImageDataGenerator(
    shear_range = 0.2, # random application of shearing
    zoom_range = 0.2,
    rotation_range = 40,  # Random rotation of images by 40 degrees
    horizontal_flip = True) # randomly flipping half of the images horizontally

test_datagen = ImageDataGenerator()

# In train_dataset and test_dataset variables, increase the training and test datasets respectively
# by applying the .flow method to train_datagen and test_datagen. The increase will be over 128 batche

train_dataset = train_datagen.flow(X_train,y_train,
                         batch_size = 128)

test_dataset = test_datagen.flow(X_test, y_test, batch_size = 128)
```

# Third part: LeNet model

```python
# Model architecture
inputs=Input(shape = (width, height, 3), name = "Input")


conv_1 = Conv2D(filters = 30,                    # filters number
                kernel_size = (5, 5),            # Size of the kernel
                padding = 'valid',               # Mode
                input_shape = (28, 28, 3),     # Dimensions of the input image
                activation = 'relu')             # activation function

max_pool_1 = MaxPooling2D(pool_size = (2, 2))

conv_2 = Conv2D(filters = 16,                    
                kernel_size = (3, 3),          
                padding = 'valid',             
                activation = 'relu')

max_pool_2 = MaxPooling2D(pool_size = (2, 2))

flatten = Flatten()

dropout = Dropout(rate = 0.2)

dense_1 = Dense(units = 128,
                activation = 'relu')

dense_2 = Dense(units = 12,
                activation = 'softmax')
x=conv_1(inputs)
x=max_pool_1(x)
x=conv_2(x)
x=max_pool_2(x)


x=dropout(x)
x=flatten(x)
x=dense_1(x)
outputs=dense_2(x)

lenet = Model(inputs = inputs, outputs = outputs)

# Compilation
lenet.compile(loss='categorical_crossentropy',  # loss function
              optimizer='adam',                 # gradient descent algorithm
              metrics=['accuracy'])             # evaluation metric
```

```python
# TimingCallback definition needed for Callbacks

class TimingCallback(Callback):
    def __init__(self, logs={}):
        self.logs=[]
    def on_epoch_begin(self, epoch, logs={}):
        self.starttime = timer()
    def on_epoch_end(self, epoch, logs={}):
        self.logs.append(timer()-self.starttime)

# Callbacks

early_stopping = EarlyStopping(
                                patience=5, # number of epochs without improvement after which the training will be stopped
                                # restore_best_weights: set this metric to True to keep the best weights once stopped
                                min_delta = 0.01, # the learning process will be stopped if the change of the 
                                                  # monitored value is less than 1%
                                verbose=1, # show the working epoch
                                mode = 'min',
                                monitor='val_loss')

reduce_learning_rate = ReduceLROnPlateau(
                                    monitor="val_loss",
                                    patience=3,
                                    episilon= 0.01, # set the threshold of variation of the metric followed with monitor
                                    factor=0.1,  # reduction factor of the learning coefficient
                                    cooldown = 4, # Number of epochs of 'pause' between two measurements of the learning rate
                                    verbose=1)

time_callback = TimingCallback()
```

```python
# Training

batch_size = 128
epochs = 30 
model_history = lenet.fit_generator(
            train_dataset, # use augmented images for train 
            steps_per_epoch=X_train.shape[0] // batch_size,
            validation_data = test_dataset, # use augmented images for test
            epochs = epochs,
            callbacks = [
                         reduce_learning_rate,
                         early_stopping,
                         time_callback
                        ],
            verbose=True)
```

```python
# Save model and history
def save_model_and_history(model, history_list, filename_sufix):
    model_filename = f'model_{filename_sufix}.h5'
    history_filename = f'histories_{filename_sufix}.pkl'
    model.save(model_filename)
    with open(history_filename, 'wb') as file:
        pickle.dump(history_list, file)

#save_model_and_history(lenet, model_history, "lenet_200x200")
save_model_and_history(lenet, model_history, "lenet_200x200_balanced")
```

```python
# Volvemos a definir el test_generator con shuffle=False.
# Esto desactiva la mezcla aleatoria de los datos en cada poca durante el proceso de evaluacin lo cual es til para asegurarte
# de que las predicciones coincidan con las etiquetas en el orden correcto.
test_dataset = test_datagen.flow(X_test, y_test, batch_size = 128, shuffle=False)


# Obtener el nmero total de muestras en el conjunto de prueba
num_samples_test = len(test_dataset) * batch_size  # 128 definido antes, ojo que en el codigo reutilizo la variable para el training
print("num_samples_test", num_samples_test)

# Obtener las etiquetas reales del conjunto de prueba
y_true = []
sample_count = 0
for _, labels in test_dataset: # al iterar sobre test_dataset se generan al mismo tiempo las imagenes aumentadas
                               # si repitiera el loop obtendria resultados diferentes
    batch_size = len(labels)
    y_true.extend(np.argmax(labels, axis=1)) # np.argmax(... recupera las etiquetas numericas desde las categoricas
                                             # (one-hot encoding)
    sample_count += batch_size
#    if sample_count >= 1648: #num_samples_test
    if sample_count >= 2744: #num_samples_test
        break
y_true = np.array(y_true)

# Realizar predicciones en el conjunto de prueba
predictions = lenet.predict(test_dataset, steps=len(test_dataset)) # aqui no se reactiva la regeneracion de imagenes
                                                                   # porque ya se generaron antes. Pero si no hubiera
                                                                   # calculado y_true previamente entonces si se generaria.
                                                                   # Si calculara y_true despues de este paso
                                                                   # se volveria a generar imagenes y 
                                                                   # se alteraria el orden y_true/predictions

# Convertir las predicciones continuas en clases
predicted_classes = np.argmax(predictions, axis=1)

print("length y_true", len(y_true))
print("length predicted_classes", len(predicted_classes))
# Mostrar el classification report
print(classification_report(y_true, predicted_classes))
```

```python
# measure accuracy
accuracy = np.mean(predicted_classes == y_true)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
# Obtener la matriz de confusin
cm = confusion_matrix(y_true, predicted_classes)

# Normalize the confusion matrix to show percentages
cm_percent = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis] * 100

#classes
classes = list(category_to_label.keys())

plt.figure()

plt.imshow(cm_percent, interpolation='nearest',cmap='Blues')
plt.title("Confusion Matrix")
plt.colorbar()
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes, rotation=90)
plt.yticks(tick_marks, classes)

for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    plt.text(j, i, f"{cm_percent[i, j]:.0f}",  # Display percentage value
             horizontalalignment="center",
             color="white" if cm_percent[i, j] > (cm_percent.max() / 2) else "black")


plt.ylabel('True labels')
plt.xlabel('Predicted Labels')
plt.show()
```

# Fifth Part: Interpret the results: Separate notebook, loading models

```python
# DONDE HAY MAXIMA CONFUSION?

from sklearn.metrics import confusion_matrix

# Obtener la matriz de confusin
cm = confusion_matrix(y_true, predicted_classes)

# Rellenar la diagonal principal con ceros para que no afecte al mximo fuera de la diagonal
np.fill_diagonal(cm, 0)

# Encontrar el valor mximo fuera de la diagonal
max_confusion = np.max(cm)

# Encontrar las clases involucradas (ndices) donde se encuentra el valor mximo
class_indices = np.where(cm == max_confusion)

# Obtener las etiquetas (nombres) correspondientes a las clases involucradas
class_labels = [list(category_to_label.keys())[list(category_to_label.values()).index(idx)] for idx in class_indices]

print("Mayor confusin:", max_confusion)
print("Clases involucradas:", class_labels)
```

La razn por la cual una clase se confunde mucho con otra puede deberse a varias razones:

    Parecido en apariencia

    Desequilibrio de clases: Si el conjunto de entrenamiento contiene una cantidad significativamente mayor de muestras de una clase que de la otra, el modelo puede volverse ms hbil en la prediccin de la clase dominante, lo que lleva a una mayor confusin entre las clases minoritarias.

    Complejidad del modelo: El modelo utilizado puede no ser lo suficientemente complejo para capturar las diferencias sutiles entre algunas clases.

    Problemas de adquisicin de datos: Si las imgenes de ciertas clases tienen condiciones de iluminacin, ngulos, tipo de maceta o resoluciones especficas que dificultan la identificacin, el modelo puede cometer ms errores en esas clases.

    Ruido en los datos: Si las etiquetas en el conjunto de entrenamiento contienen errores o estn mal etiquetadas, el modelo puede aprender incorrectamente y cometer confusiones entre clases similares.

```python
# CREACION DE MASCARAS PARA ELIMINAR PIXELES NO MUY VERDES (TIERRA) O BIEN PIXELES MUY BLANCOS (MACETAS, ETIQUETAS, ETC)
import random
import matplotlib.pyplot as plt

# Etiqueta especfica que quieres mostrar en el mosaico
target_label = 7
faktor_green = 1.4
threshold_white =150

# Obtener el nombre de la etiqueta especfica usando el mapeo inverso category_to_label
category_name = list(category_to_label.keys())[list(category_to_label.values()).index(target_label)]

# Filtrar las imgenes que corresponden a la etiqueta target_label
target_images = X[Y_labels == target_label]

# Verificar si hay suficientes imgenes con la etiqueta target_label
if len(target_images) >= 9:
    # Tomar 9 imgenes al azar (si hay ms de 9)
    target_images = random.sample(list(target_images), 9)
else:
    # Si hay menos de 9 imgenes, llenar con imgenes en blanco
    num_missing_images = 9 - len(target_images)
    blank_image = np.zeros((width, height, 3), dtype=np.uint8)  # Crear una imagen en blanco
    target_images = target_images.tolist()
    target_images.extend([blank_image] * num_missing_images)

# Crear el mosaico de 3x3
fig, axes = plt.subplots(3, 3, figsize=(10, 10))

# Mostrar todos los canales en el mosaico con la mscara aplicada
for i in range(3):
    for j in range(3):
        # Obtener la imagen original de la clase
        image = target_images[i * 3 + j]
        
        # Aplicar la mscara para filtrar colores no muy verdes y los blancos
        mask_green = (image[:, :, 1] < faktor_green * image[:, :, 0]) & (image[:, :, 1] < faktor_green * image[:, :, 2])
        mask_white = (image[:, :, 0] > threshold_white) & (image[:, :, 1] > threshold_white) & (image[:, :, 2] > threshold_white)     
        mask = mask_green | mask_white  # Combinar las mscaras utilizando OR
        image[mask] = [0, 0, 0]  # Convertir a negro aquellos pxeles donde verde es dbil o blanco es intenso

        
        # Agregar ttulo a cada subfigura con el nmero de imagen y el nombre de la clase
        image_number = i * 3 + j + 1
        axes[i, j].imshow(image)
        axes[i, j].set_title(f"Image {image_number}")
        axes[i, j].axis('off')

# Agregar ttulo al mosaico con el nombre de la clase
fig.suptitle(f"Mosaico imgenes filtradas para la etiqueta: {category_name}", fontsize=16)

plt.show()
```
