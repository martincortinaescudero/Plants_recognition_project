```python
#Import
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file and the specific sheet
df = pd.read_excel("Education.xlsx", sheet_name="Education")

df['USA (School)'] = df['USA (School)'] * 100
df['Germany (School)'] = df['Germany (School)'] * 100
df['Mexico (School)'] = df['Mexico (School)'] * 100

#Display
df.head()
```

**SCHOOL**

```python
# Plotting the data
plt.figure(figsize=(12, 5))

# Plot each country
plt.plot(df['Year'], df['USA (School)'], label='United States')
plt.plot(df['Year'], df['Germany (School)'], label='Germany')
plt.plot(df['Year'], df['Mexico (School)'], label='Mexico')

# Adding titles and labels
plt.title('Primary and High school')
plt.xlabel('Year')
plt.ylabel('Enrollment (%)')

# Adding a legend
plt.legend()

# Display the plot
plt.grid(False)
plt.show()
```

**UNIVERSITY**

```python
# Plotting the data
plt.figure(figsize=(12, 5))

# Plot each country
plt.plot(df['Year'], df['USA (university)'], label='United States')
plt.plot(df['Year'], df['Germany (university)'], label='Germany')
plt.plot(df['Year'], df['Mexico - Both genders (university)'], label='Mexico')

# Adding titles and labels
plt.title('University Attendance')
plt.xlabel('Year')
plt.ylabel('Enrollment (%)')

# Adding a legend
plt.legend()

# Display the plot
plt.grid(False)
plt.show()
```
